```python
# !pip install bs4
# !pip install lxml
# !pip install frozendict
# !pip install iteration_utilities
# !pip install selenium
# !pip install webdriver_manager
import pandas as pd
import time
import csv
import scrapy
from scrapy.crawler import CrawlerProcess
from urllib import response
```

```python
with open("all_hrefs.txt", "r", newline="\n") as f:
    all_hrefs = eval(f.read())
```

```python
curses = []

class CursesSpider(scrapy.Spider):
    name = "CursesSpider"
    start_urls = all_hrefs
    custom_settings = {
        'DOWNLOAD_DELAY': 0.35 
        }
    

    def parse(self, response):
        curse = {
            "name": response.xpath('//h1[@class="text-h2-bold q-mb-xs lt-md"]/text()').get(),
            "price": response.xpath('//div[@class="col-md-8 col-12 order-2"]/div[@class="text-h0-bold"]/span/text()').get().replace('\xa0',''),
            "ratio": response.xpath('//div[@class="text-h4 l-course__position"]/text()').get(),
            "time": response.xpath('//div[@class="flex items-center"]/text()').get(),
            "skills": response.xpath('//div[@class="l-skills__items"]//div[@class="l-skills__item-name"]/text()').getall() + response.xpath('//div[@class="l-skills__items"]//a[@class="l-skills__item-name"]/text()').getall(),
            "school": response.xpath('//a[@class="owners-list__info-title"]/text()').get()
             }
        curse.update(dict(zip([title.strip() for title in response.xpath('//div[@class="l-features__title"]/text()').getall()],
                              [title.strip() for title in response.xpath('//div[@class="l-features__value"]/text()').getall()])))
        curses.append(curse)
                    
        

process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(CursesSpider)
process.start()
```

```python
pd.DataFrame(curses).to_csv('curses.csv')
```
